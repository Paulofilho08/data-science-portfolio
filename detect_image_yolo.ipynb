```python
from ultralytics import YOLO

# Carregar modelo pré-treinado (YOLOv8 nano)
model = YOLO("yolov8n.pt")

# Fazer predição em uma imagem
results = model.predict(source="/content/img_5837-jjkmghdktx.jpg", conf=0.5)
results[0].show()  # Mostra imagem com detecções
```
